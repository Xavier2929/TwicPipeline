# first I'll see if I can get a list of the latest twic


```python
import requests
import zipfile
import os
from io import BytesIO
from bs4 import BeautifulSoup
import re
```

```python

twic_url = os.getenv("TWIC_URL")
twic_file_folder = os.getenv("TWIC_FILES_FOLDER")
file_type_request = ".zip"
twic_postfix = "g"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0 Safari/537.36",
    "Accept": "application/json,text/html;q=0.9,*/*;q=0.8"
}
```

<h3> Get the latest twic number </h3>

```python
twic_main_site = os.getenv("TWIC_MAIN_SITE")
twic_main_site_request = requests.get(twic_main_site,headers=headers)

soup = BeautifulSoup(twic_main_site_request.content, "html.parser")
text = soup.get_text(" ", strip=True)
match = re.search(r"The Week in Chess\s+(\d+)", text)
latest_issue = ""

if match:
    latest_issue = (match.group(1))
    print("Latest issue:", latest_issue)
else:
    print("No se pudo encontrar el latest issue")
```

```python
twic_request_str = twic_url+latest_issue+twic_postfix+file_type_request
```

```python
twic_request = requests.get(twic_request_str,headers = headers)
twic_result = twic_request.json
twic_status_code = twic_request.status_code
twic_filetype = twic_request.headers.get("Content-Type")
```

```python
print(f'the result of the request is {twic_status_code}')
print(f'the filetype of the request is {twic_filetype}')
```

```python
# build the path for the file
twic_filepath = twic_file_folder+"/"
twic_filepath
```

```python
zip_bytes = BytesIO(twic_request.content)
```

```python

with zipfile.ZipFile(zip_bytes) as zip_file:
    for name in zip_file.namelist():
        with zip_file.open(name) as file:
            content = file.read()
            file_path = twic_filepath + name
            print(file_path)
            with open(file_path, "wb") as f:
                f.write(content)
```
